```python
def crop64(DIR):
    image = imread(DIR)
    #plt.imshow(image)
    x, y, _ = image.shape
    x = int(x / 2)
    y = int(y / 2)
    image = image[(x - 64):(x + 64), (y - 64):(y + 64), :]
    image = resize(image, output_shape=(64, 64), mode='reflect')
    return image
```
